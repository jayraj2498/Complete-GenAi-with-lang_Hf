```python
# Here we load the data and do chunking -> embedding --.> then store embedding into vectoreDB 

from langchain_community.document_loaders import TextLoader 
loader = TextLoader("speech.txt") 
docs = loader.load() 
docs 
```

```python
# we split data into chunks 

from langchain_text_splitters import RecursiveCharacterTextSplitter 

text_splitter=RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50) 
final_documents = text_splitter.split_documents(docs)
final_documents
```

```python
# Now we do embedding of this chunks and store it into vectoreDB 

from langchain_community.vectorstores import Chroma 

db = Chroma.from_documents(final_documents , embeddings) 
db 
```

```python
# Now we can retrive query the result from vectoreDB 

our_query = "The world must be made safe for What ?"

retrive_result =db.similarity_search(our_query) 

print(retrive_result)
```

```python
### Retrieve the results from query vectorstore db
query="It will be all the easier for us to conduct ourselves as belligerents"
retrieved_results=db.similarity_search(query)
print(retrieved_results)
```
